1) Abrir el archivo "Emisiones_CO2.csv" y cargar sus datos en un diccionario.

```python
import os
archivo = open('Emisiones_CO2.csv', 'r')
```

```python
dicc_emisiones = {  'cod_pais' : [], # 0
                    'nom_pais' : [],  #1
                    'region' : [],  #2 
                    'anio' : [], #3
                    'co2' : [], #4
                    'co2_percapita' : []}#5
```

```python
next(archivo)
totalco2 = 0
for linea in archivo: 
    linea = linea.split(sep = "|")
    for a, i in enumerate(linea):
        i = i.replace("." , "")
        i = i.replace("," , ".")
        linea [a] = i
        if linea[a] == "" or linea [a] == "\n":
            linea[a] ="valor no encontrado"
    dicc_emisiones["cod_pais"].append(linea[0])
    dicc_emisiones['nom_pais'].append(linea[1])
    dicc_emisiones['region'].append(linea[2])
    dicc_emisiones['anio'].append(int(linea[3]))
    if linea [4] == "valor no encontrado":
        dicc_emisiones['co2'].append(linea[4])
    else:
        dicc_emisiones['co2'].append(float(linea[4]))
    if linea [5] == "valor no encontrado":
        dicc_emisiones['co2'].append(linea[5])
    else:
        dicc_emisiones['co2'].append(float(linea[5]))

    if linea[2]=="Am√©rica Latina y Caribe" and linea[3]=="2010":
        if linea[4]=="valor no encontrado":
            continue
        else:
            totalco2 +=float(linea[4])

print(dicc_emisiones)
print(dicc_emisiones["cod_pais"])
print(dicc_emisiones["nom_pais"])
print(dicc_emisiones["region"])
print(dicc_emisiones["anio"])
print(dicc_emisiones["co2"])
print(dicc_emisiones["co2_percapita"])
print(totalco2)
columna1 = len(dicc_emisiones["cod_pais"])
columna2 =len(dicc_emisiones["nom_pais"])
columna3 =len(dicc_emisiones["region"])
columna4 =len(dicc_emisiones["anio"])
columna5 =len(dicc_emisiones["co2"])
columna6 =len(dicc_emisiones["co2_percapita"])
```

```python
region = dicc_emisiones['region']
anio = dicc_emisiones['anio']
co2 = dicc_emisiones['co2']

total = []
lista_floar = []
for a,b,c in zip(region,anio,co2):
    if a == 'Am√©rica Latina y Caribe' and b == '2010':
        total.append(c.replace('.','').replace(',','.'))



total = [x for x in total if x != '']

total = [float(x) for x in total]

print(total)
```
