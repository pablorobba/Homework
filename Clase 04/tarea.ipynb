```python
import pandas as pd
df = pd.read_csv("hospitales2.csv",delimiter= ",")
df.head()
```

```python
nuevo_archivo = df.copy()
nuevo_archivo["WKT"]= nuevo_archivo.WKT.str.replace("POINT","")
nuevo_archivo["WKT"]= nuevo_archivo.WKT.str.replace("(","")
nuevo_archivo["WKT"]= nuevo_archivo.WKT.str.replace(")","")

latitud_longitud = nuevo_archivo["WKT"]

latitud_longitud = latitud_longitud.str.split(" ",expand= True)
longitud,latitud = latitud_longitud[1],latitud_longitud[2]

name = nuevo_archivo["DOM_NORMA"]
label = nuevo_archivo["NOM_MAP"]
name
```

```python
nuevo_CSV = pd.DataFrame(columns=["latitude","longitude","name","label"])
nuevo_CSV["latitude"] = latitud
nuevo_CSV["longitude"] = longitud
nuevo_CSV["name"] = name
nuevo_CSV["label"] = label
nuevo_CSV
```

```python
nuevo_CSV.to_csv("hospitales3.csv",index= False)
```
